```python
le = preprocessing.LabelEncoder()
le.fit(dataframe["region"])
dataframe["region_encoded"] = le.transform(dataframe["region"])
sb.pairplot(dataframe.dropna(), hue='type',height=4,vars=["AveragePrice","region_encoded","Unnamed: 0"],kind='scatter')
```

```python

X = np.array(dataframe[["AveragePrice","region_encoded","Unnamed: 0"]])

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
y = np.array(dataframe['year'])
X.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d')
colores=['blue','red','pink','orange']
asignar=[]

le = preprocessing.LabelEncoder()
le.fit(y)
y = le.transform(y)
for row in y:
    asignar.append(colores[row])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
```

 3. Determina un valor de k.

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

Con la gráfica anterior decidimos que el valor de K sea 4 porque es nuestro "punto de codo".

4. Utilizando scikitlearn calcula los centros del algoritmo k-means.

```python
kmeans = KMeans(n_clusters=4).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
dist_c = np.zeros((len(centroids),len(centroids)))
for i,c1 in enumerate(centroids):
  for j,c2 in enumerate(centroids):
    dist_c[i,j] = np.sqrt((c1[0]-c2[0])**2+(c1[1]-c2[1])**2+(c1[2]-c2[2])**2)
print(dist_c)
```

```python
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow','pink','purple','black','orange','brown']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores[:len(centroids)], s=1000)
```
