# Obtencion de la media, mediana y desviacion estandar

```python
print("La media del precio promedio es: " + str(data['AveragePrice'].mean()))
print("La mediana del precio promedio es: " + str(data['AveragePrice'].median()))
print("La desviacion estandar del precio promedio es: " + str(data['AveragePrice'].std()))
```

```python
print("La media del precio promedio es: " + str(data['Total Volume'].mean()))
print("La mediana del precio promedio es: " + str(data['Total Volume'].median()))
print("La desviacion estandar del precio promedio es: " + str(data['Total Volume'].std()))
```

# Análisis

#### El conjunto de datos consta de 18249 observaciones de 14 columnas.

### Significado de las variables:
 - Date: fecha de la observación 
 - Average Price: precio promedio de un solo aguacate tipo orgánico o convencional
 - Type: convencional u orgánico
 - Year: año
 - Region: ciudad o región de observación
 - Total Volume: número total de aguacates vendidio
 - (4046, 4225, 4770): PLU -> código identificar a los productos frescos o de granel, basándose en distintos aspectos como: tipo de mercancía, variedad, tamaño y tipo de cultivo y a veces el lugar de origen.

#### Un poco más del 2% de los aguacates vendidos son de tipo orgánico. Por lo tanto el tipo convencional tiene más demanda.

#### Como podemos ver en el mapa de calor anterior, todas las características no están correlacionadas con la columna de precio promedio, sino que la mayoría de ellas están correlacionadas entre sí. 

#### Como se puede ver en el histograma, la distribución del precio medio muestra que, en la mayoría de los casos, el precio del aguacate está entre 1,1 y 1,4.
